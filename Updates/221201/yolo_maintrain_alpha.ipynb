```python
#라벨 경로를 전달받아서, add_num 만큼 라벨의 첫번째 문자열을 증가시켜주는 함수 
def change_label(path, add_num):
  total_str = ""

  with open(path, 'r') as f:
    data = f.read()
  if len(data) > 3:
    split_data = data.split("\n")
    #print('split_data = ', split_data)

    #print('\n')
    for f_data in split_data:
      #print("f_data = ", f_data)
    
      
      f2_data = f_data.split(" ")
      f2_data[0] = str(int(f2_data[0]) + add_num)
      #print('f2_data = ', f2_data)
      
      res = ""
      for s in f2_data:
        res += s + ' '

      res = res.strip()

      total_str += res + '\n'
      #print("total_str = ", total_str)

    with open(path, 'w') as f:
      f.write(total_str.strip())
    
    #print('success')
```
